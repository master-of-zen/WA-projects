# Idea
Exploring current state of HP/Man balance in WA for divisions and support companies specifically.
What I want to show on this data that in current state support companies and frontline battalionss have large amount of Man and equipment, with near 0 HP stat to represent it as a part of the division, which leads to situation where support companies (later on some frontline battalions) make division more fragile, and by a lot than without them.

### Divisions
In this example so far I have used GER division at time of 1 SEP 39, and they represent whole force of GER at that point.

### IC/HP
How many IC in division are represented by 1 HP.

### Manpower/HP
How many people in division are represented by 1 HP.

### IC/HP%
Difference beetween division with and without support companies.
Reads as "By adding/removing support companies, I spend/lose X % more IC per HP than without them"

### Manpower/HP%
Difference beetween division with and without support companies.
Reads as "By adding/removing support companies, I spend/lose X % more IC per HP than without them"


```python
# This code provides some analisis of current WA balance regarding divisions HP, IC, Manpower

from __future__ import division
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


# Data is taken from templates at 1 sep 1939
divisions = [
  [ "Name", "HP", "IC", "Man"],
  ["GER INF 39 ", 201.7, 2932, 16700],
  ["GER INF no support 39", 197.6, 2047, 13000],
  ["GER Mont 39", 204.8, 2668, 15100],
  ["GER Mont no support 39", 201.1, 1851, 12000],
  ["GER Armor 39", 188.2, 13443,15500],
  ["GER Armor no support 39", 181, 10967, 11000],
  ["SOV INF 39", 177.9, 3485, 16700],
  ["SOV INF no support 39", 172.1, 2129, 13000],
  ["SOV Mont 39", 176.4, 1770, 13700],
  ["SOV Mont no support 39", 174.8, 1968, 10500],
  ["SOV Moto 39", 209.6, 4192, 14900],
  ["SOV Moto no support 39", 206.7, 3604, 13300],
  ["CHI INF 39", 116.3, 1168, 10700], # IT'S 12.8 width
  ["CHI INF no support 39", 112.0, 672, 8000],
  ["JAP INF", 192.6, 2209, 17100],
  ["JAP INF no support 39",187.8, 1476, 14000],
  ["ENG INF", 222.6, 3184,17200],
  ["ENG INF no support", 215.6, 1749,13000],
  ["ENG Armor 39", 163.3, 8132,14900],
  ["ENG Armor no support 39", 159, 6379, 11000],
  ["GER Armor 43", 194.7, 14051, 15500],
  ["GER Armor no support 43", 187.5, 11395, 11000],
  ["SOV INF 43", 221.8, 3022, 18700],
  ["SOV INF no support 43", 216.0, 1833, 15000],
  ["SOV Armor Med 43", 208, 10290, 16800],
  ["SOV Armor Med no support 43", 201.6, 9015, 13200],
  ["SOV Armor Med 43", 208, 10290, 16800],
  ["SOV Armor Med no support 43", 201.6, 9015, 13200],
  ["US mech", 322, 5880, 19100],
  ["US mech", 322, 4331, 15000]

]


## Do calculations
divisions[0].extend(["IC/HP","Man/HP"])
for i, division in enumerate(divisions):
  if i == 0:
    continue

  # Calculate IC/HP
  division.append(round(division[2]/division[1], 1))

  # Calculate Manpower/HP
  division.append(round(division[3]/division[1], 1))


# Do percents
divisions[0].extend(["IC/HP %", "Man/HP%"])
for i, division in enumerate(divisions):
  if i == 0:
    continue 

  # Calculate IC/HP %
  if i%2 ==0:
    division.append(round((divisions[i][4]/divisions[i-1][4]) * 100 - 100, 1))
  else:
    division.append(round((divisions[i][4]/divisions[i+1][4]) * 100 - 100, 1))
  
  

  # Calculate Manpower/HP %
  if i%2 ==0:
    division.append(round((divisions[i][5]/divisions[i-1][5]) * 100 - 100, 1))
  else:
    division.append(round((divisions[i][5]/divisions[i+1][5]) * 100 - 100, 1))


df = pd.DataFrame(divisions[1:], columns=divisions[0])

by_ic = df.sort_values(by='IC/HP %',ascending=False)
by_hp = df.sort_values(by='Man/HP%',ascending=False)

df
```

# Problem

Not having HP that represent added by support company or battalion manpower and equipment have next consecuences.

## Division size

HP that bat/com lacking need to be compensated, in this case by infantry battalions, so in order to have a division that is strong enough to see combat and don't have enourmous casualties in manpower and equipment it's highly rewarding to have as many of them as possible to offset no HP. 
Which leads to any division width below 30 with support companies being a lot weaker, or smaller divisions to have less of them, probably even bare minimum of couple of them.

Fixed size support companies provide divisions with large amount of menpower at around 25-35% with near zero HP increase.

## Optimal division

As WA stands currently, best strength gained is from infantry, moto, mech, and mechanized divisions being extremely resilent to damage in comparison to any other type of division, and being able to use that HP pool to compensate for filling all support company slots.

## Abnormal casualties for high cost/ high investment divisions

High cost divisions as Armor divisions for each country, require a lot of equipment and IC investement, and it's expected from divison to be a highly active unit. Armor division have really hight Manpower/HP count, mainly located in support companies with near 0 HP, and IC/HP being a ~6 times of infantry division.

It's expected that IC/HP of armor division would be higher as more costly equipment is used, given that, receiving 6 times the cost of 1 HP damage likely to be too much.

Armor divisions also being ones that have higher recovery rate and have combat more frequently leads to them receiving abnormal amount of damage in both manpower and equipment.


# Expected results
With addition of support companies, total Manpower/HP, and IC/HP to change by less than 10%. Which will mean corresponding increase in strength of a division with additional companies.
In addition, that could be applied to frontline artillery, anti-tank battalions.

## HP calculation
HOI4 base game balance intends that support and frontline battalions that are not infantry have lower HP, and lower division sizes still suffer disproportionate casualties. Also Vanilla have less of them and lacks in variability of battalions and companies.
New HP formula would be lower or comparable to light infantry HP per 100 man.
For test 1.2 HP per 100 without doctrines + some amount of equipment HP.

Which will make most of support companies to have 3-9 HP.
